```python
import pandas as pd 
import numpy as np 
data = pd.read_csv('Data/sales_2019.csv')
data.head()
```

```python
data['Month'] = data['Order Date'].str[:2]
data.head()
```

```python
data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head()
```

# Thang nao ban duoc nhieu nhat va thang do doanh thu duoc bao nhieu ? 

```python
import matplotlib.pyplot as plt 
sales = data.groupby('Month')[['Quantity Ordered', 'Price Each', 'Sales']].sum()
sales
```

```python
plt.figure(figsize = (20,5))
plt.bar(range(1,13), sales['Sales'])
plt.title('Sales per month')
plt.xticks(range(1,13))
plt.xlabel('Month')
plt.ylabel('Sales in USD')
```

# Thanh pho nao ban duoc nhieu san pham nhat 

```python
data.head()
```

```python
data['Purchase Address'].str.split(',').str[1].str.strip()
```

```python
data['City'] = data['Purchase Address'].str.split(',').str[1].str.strip()
```

```python
data.head()
```

```python
Highest_quantity = data.groupby('City')['Quantity Ordered'].sum()
```

```python
Highest_quantity = pd.DataFrame(Highest_quantity)
Highest_quantity = Highest_quantity.reset_index(names= 'City')
```

```python
Highest_quantity
```

```python
Highest_quantity[Highest_quantity['Quantity Ordered'] == Highest_quantity['Quantity Ordered'].max()]['City']
```

```python
plt.figure(figsize=(20,5))
plt.bar(Highest_quantity['City'], Highest_quantity['Quantity Ordered'])
plt.title('Quantity per city')
plt.xlabel('City')
plt.ylabel('Quantity')
```

```python
highest_sales = data.groupby('City')[['Quantity Ordered', 'Sales', 'Price Each']].sum()
highest_sales = highest_sales.reset_index()
```

```python
highest_sales
```

```python
plt.figure(figsize=(20,4))
plt.bar(highest_sales['City'], highest_sales['Sales'])
plt.title('Sales per City')
plt.xticks(highest_sales['City'], rotation = 'vertical') #Label truc X thanh nam doc
plt.xlabel('City')
plt.ylabel('Sales')
```

# Thoi gian nao ta nen hien thi san pham la hop ly nhat de khach hang co the mua 


```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.head()
```

```python
data['Hour'] = data['Order Date'].dt.hour
data['Minute'] = data['Order Date'].dt.minute
data.head()
```

```python
hour = [hr for hr,data in data.groupby('Hour')] #data.groupby(hour) tra ve cac cap (key, sub_dataframe) -> hr chi lay phan key
plt.figure(figsize=(20,5))
plt.plot(hour,data.groupby('Hour')['Quantity Ordered'].count(), color = 'c')
plt.xticks(hour)
plt.grid() #Tao luoi
plt.title('Time of the Day when sales are more')
plt.xlabel('Time (hour)')
plt.ylabel('Order count')
```

=> trong khoang 11am -> 1pm va 6pm ->8pm
